```python
x_values = []
y_values = []
with open('test.txt', 'r') as file:
    for line in file:
        x, y = map(int, line.split())
        x_values.append(x)
        y_values.append(y)
plt.plot(x_values, y_values, marker='o')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Line Plot from test.txt Data')
plt.show()
```
