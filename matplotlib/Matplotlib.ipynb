# 2D Line Plot

- Bivariate Analysis - Analysis in 2 column
- Categorical -> numerical and numerical ->numerical
- Use case - Time Series data

```python
price = [48000,54000,57000,47000,45000]
year = [2015,2016,2017,2018,2020]
plt.plot(year,price)
plt.show()
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.grid()
plt.plot(batsman['index'],batsman['V Kohli'])
plt.show()
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
# label title
plt.grid()
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# color(hex) and line(width and style) marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange')

plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# linestyle solid ,dashed,dotted
# linewidth
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth =3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange',linestyle='dashdot',linewidth=2)

plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# marker "D"-Diamond, 'o'-point,'^'-Triangle
#markersize = 10
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth =3,marker='.')
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange',linestyle='dashdot',linewidth=2,marker='o')

plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# label if label is used then call legend
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth =3,marker='.',label ='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend(loc='best')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.show()
plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2015,2019)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth =3,marker='.',label ='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange',linestyle='dashdot',linewidth=2,marker='.',label='Rohit')

plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
# show- plt.show() is the main function to show the graph in other code editor like pycharm, vs code we have to use plt.show
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth =3,marker='.',label ='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange',linestyle='dashdot',linewidth=2,marker='.',label='Rohit')

plt.title("Rohit Sharma Vs Virat Kohli Career Comparision")
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

# Scatter Plot

- Bivariate Analysis
- Numerical vs Numerical
- use case - finding correlation

```python
# plot using sacteer plot
# genereate random no.
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df=df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='orange',marker='*')
plt.title('Avg and SR of top 50 batsman')
plt.xlabel('Avg')
plt.ylabel('SR')
```

```python
# size
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot using plt.scatter()
# slow
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# sactter plot using plt.plot()
# faster
plt.plot(tips['total_bill'],tips['tip'],'*')
```

# Bar Chart
- Bivariate Analysis
- Numerical vs Categorical
- use case - Aggregate analysis of groups

# y - aggregate  like frequency, avg_age
# x- category

```python
# Simple bar chart
children = [6,8,5,2,9]
colour =['Red','Blue','Green','Pink','Yellow']
plt.bar(colour,children,color='orange')
plt.xlabel('colour')
plt.ylabel('frequency of student favourite colour')
```

```python
# Simple bar chart
children = [6,8,5,2,9]
colour =['Red','Blue','Green','Pink','Yellow']
plt.barh(colour,children,color='orange')
plt.ylabel('colour')
plt.xlabel('frequency of student favourite colour')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

# using pandas plot

```python
df.plot(kind='bar',x='batsman')
plt.xlabel('Batsman')
plt.ylabel('Run')
plt.show()
```

```python
print(df.shape[0])
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color ='Red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='Yellow')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='Blue')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# Simple bar chart
children = [6,8,5,2,9]
colour =['Red Red Red Red Red','Blue Blue Blue Blue Blue','Green Green Green Green Green','Pink Pink Pink Pink Pink','Yellow Yellow Yellow Yellow Yellow']
plt.bar(colour,children)
plt.xlabel('colour')
plt.ylabel('frequency of student favourite colour')
plt.xticks(rotation=60)
plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')
```

# Pie chart
- Univariate/Bivariate
- Categorical vs Numerical
- Use case - To find contribution of standard scale

```python
# Simple data
data =[23,45,100,20,49]
sub=['Eng','Sci','maths','sst','hindi']
plt.pie(data,labels=sub)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# use autopct - %0.1f%%
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#pass color
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','brown','white'])
plt.show()
```

```python
# pass explode the particular component pass intput array at distance radius [1 or 5]etc.
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','brown','grey'],explode=[0,0,0,2,0,0])
plt.show()
```

```python
# showdow give 3-D observation
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','brown','white'],explode=[0,0,0,0.1,0,0],shadow=True)
plt.show()
```
